```python
import json
from kafka import KafkaConsumer
from aiogram import Bot, Dispatcher, types
from aiogram.filters.command import Command
import asyncio
from aiogram.enums import ParseMode
```

```python
SERVER = "broker:9092"
consumer = KafkaConsumer('streaming', bootstrap_servers=[SERVER])
```

```python
bot = Bot(token="7133816981:AAEBpw-7Dt5DkYFKAzpcWr4al2ULJq_Q_Js")
dp = Dispatcher()

@dp.message(Command("start"))
async def cmd_start(message: types.Message):
    for km in consumer:
        offer_info = json.loads(km.value.decode('utf-8'))
        for k, v in offer_info.items():
            if v == None:
                v = "Not provided"
        text = f"""<b>Job title:</b> {offer_info['jobTitle']}
<b>Employer:</b> {offer_info['employerName']}
<b>Location:</b> {offer_info['locationName']}
<b>Minimum salary:</b> {offer_info['minimumSalary']}
<b>Applications:</b> {offer_info['applications']}
<b>URL:</b> {offer_info['jobUrl']}
"""
        print(text)
        await message.answer(text, parse_mode=ParseMode.HTML)

async def main():
    await dp.start_polling(bot)

await main()
```
